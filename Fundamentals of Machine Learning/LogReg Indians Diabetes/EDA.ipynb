```python
import numpy as np
import pandas as pd
```

## Загрузка данных

* Концентрация глюкозы в плазме через 2 часа при оральном тесте на толерантность к глюкозе.
* Диастолическое артериальное давление (мм рт. Ст.).
* Толщина трехглавой кожной складки (мм).
* 2-часовой сывороточный инсулин (мю Ед / мл).
* Индекс массы тела (вес в кг / (рост в м) ^ 2).
* Родословная функция диабета.
* Возраст (годы).
* Переменная класса (0 или 1) - приобретет ли человек диабет в ближайшие 5 лет.

## Анализ исходных данных

```python
fig, ax = plt.subplots()
diabetes_df.Pegnancy_count.value_counts().sort_index().plot(kind='bar', ax=ax)
ax.set_title('Беременность')
ax.set_xlabel('Количество беременностей')
ax.set_ylabel('Количество людей')
```

Работать с нулями у этого признака не нужно, так как индейцы могли быть как мужчинами, так и женщинами, которые еще не рожали

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
diabetes_df.Glucose.hist(ax=axes[0])
axes[0].set_title("Распределение концентрации глюкозы")
axes[1].boxplot(diabetes_df.Glucose)
axes[1].set_title("Выбросы для концентрации глюкозы")

plt.show()
```

У признакак концентрации глюкозы имеются подозрительные значения в районе 0. Ящик с усами также свидетельствуют о том, что это нетипичные значения - выбросы, или искусственно замещенные пропуски. Поэтому эти значения мы заменим на медиану.

```python
median = diabetes_df.Glucose.median()
diabetes_df.Glucose = diabetes_df.Glucose.apply(lambda x: x if x != 0 else median)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
diabetes_df.Glucose.hist(ax=axes[0])
axes[0].set_title("Распределение концентрации глюкозы")
axes[1].boxplot(diabetes_df.Glucose)
axes[1].set_title("Выбросы для концентрации глюкозы")

plt.show()
```

Теперь можно идти дальше

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
diabetes_df.Pressure.hist(ax=axes[0])
axes[0].set_title("Распределение давления ")
axes[1].boxplot(diabetes_df.Pressure)
axes[1].set_title("Выбросы для давления")

plt.show()
```

У признака диастолического артериального давления также видны выбросы, причем по обоим сторонам. Пока что заменим только нули на медианные значения.

```python
median = diabetes_df.Pressure.median()
diabetes_df.Pressure = diabetes_df.Pressure.apply(lambda x: x if x != 0 else median)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
diabetes_df.Pressure.hist(ax=axes[0])
axes[0].set_title("Распределение давления ")
axes[1].boxplot(diabetes_df.Pressure)
axes[1].set_title("Выбросы для давления")

plt.show()
```

```python
(diabetes_df.Pressure > diabetes_df.Pressure.mean() + 3*diabetes_df.Pressure.std()).sum()
```

```python
(diabetes_df.Pressure < diabetes_df.Pressure.mean() - 3*diabetes_df.Pressure.std()).sum()
```

Данных, выходящих за пределы 3 сигм всего 8, поэтому можем исключить их из датасета, не боясь, что потеряем много информации.

```python
# Оставим только объекты, признак давления которых находится в пределах 3 сигм отклонения от среднего значения
mean = diabetes_df.Pressure.mean() 
std = diabetes_df.Pressure.std() 
diabetes_df = diabetes_df[(diabetes_df.Pressure < mean + 3*std) & (diabetes_df.Pressure > mean - 3*std)]
diabetes_df.index = range(len(diabetes_df))
```

После этих изменений можно двигаться дальше

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
diabetes_df.Thickness_of_skin_fold.hist(ax=axes[0])
axes[0].set_title("Распределение толщины трехглавой кожной складки")
axes[1].boxplot(diabetes_df.Thickness_of_skin_fold)
axes[1].set_title("Выбросы для толщины трехглавой кожной складки")

plt.show()
```

Имеется выброс со значением 99, его нужно будет убрать из датасета. Также есть подозрение, что в этом признаке было большое число пропущенных значений

```python
diabetes_df.Thickness_of_skin_fold.value_counts().sort_index().head()
```

Заметим, что минимальным значением без нуля является 7 мм и его имеет всего 2 человека из 700+, то есть более вероятней, что 225 значения в датасете были пропущены и их заменили нулями

```python
# Удаляем выброс
diabetes_df = diabetes_df[diabetes_df.Thickness_of_skin_fold != 99]

# Получаем ненулевые значения
thickness_without_0 = np.array([value for value in diabetes_df.Thickness_of_skin_fold.values if value != 0])

# Находим медианное значение
thickness_median = np.median(thickness_without_0)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
axes[0].hist(thickness_without_0)
axes[0].set_title("Распределение толщины трехглавой кожной складки без нулей")
axes[1].boxplot(thickness_without_0)
axes[1].set_title("Выбросы для толщины трехглавой кожной складки без нулей")

plt.show()
```

Оставим этот признак в таком виде и вернемся к нему позже, чтобы заменить нулевые значения

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
diabetes_df.Insulin.hist(ax=axes[0])
axes[0].set_title("Распределение инсулина")
axes[1].boxplot(diabetes_df.Insulin)
axes[1].set_title("Выбросы для инсулина")

plt.show()
```

По аналогии с предыдущим признаком мы также видим, что имеется большое количество пропущенных данных и несколько значений, которые отличаются от остальных сильнее всего, выберем в качестве порога отсечение значение 350.

```python
# Удаляем выбросы
diabetes_df = diabetes_df[diabetes_df.Insulin < 350]

# Получаем ненулевые значения
insulin_without_0 = np.array([value for value in diabetes_df.Insulin.values if value != 0])

# Находим медианное значение
insulin_median = np.median(insulin_without_0)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
axes[0].hist(insulin_without_0)
axes[0].set_title("Распределение инсулина")
axes[1].boxplot(insulin_without_0)
axes[1].set_title("Выбросы для инсулина")

plt.show()
```

Оставим этот признак в таком виде и вернемся к нему позже, чтобы заменить нулевые значения

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
diabetes_df.BMI.hist(ax=axes[0])
axes[0].set_title("Распределение индекса массы тела")
axes[1].boxplot(diabetes_df.BMI)
axes[1].set_title("Выбросы для индекса массы тела")

plt.show()
```

Ярко выражены выбросы в виде 0 и значения превышающие 50

```python
# Удаляем выбросы
diabetes_df = diabetes_df[diabetes_df.BMI != 0]
diabetes_df = diabetes_df[diabetes_df.BMI < 50]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
diabetes_df.BMI.hist(ax=axes[0])
axes[0].set_title("Распределение индекса массы тела")
axes[1].boxplot(diabetes_df.BMI)
axes[1].set_title("Выбросы для индекса массы тела")

plt.show()
```

Вернемся к признакам, которые мы откладывали

Их количество не сильно уменьшилось после удаления выбросов. Терять минимум 318 записей это очень дорого, поэтому заменим все нули на медианные значения, найденные на безнулевых данных

```python
diabetes_df.Thickness_of_skin_fold = diabetes_df.Thickness_of_skin_fold.apply(lambda x: x if x != 0 else thickness_median)
diabetes_df.Insulin = diabetes_df.Insulin.apply(lambda x: x if x != 0 else insulin_median)
```
