Цель задания - исследование функции при различных параметрах

$$f(x)=\frac{x^\beta+\alpha^\beta}{x^\beta} $$

Для каждой задачи необходимо:
* Построить график (размер графика должен быть достаточным, чтобы визуально увидеть особенности изучаемых функций), график каждой функции должен быть одного цвета для одного значения параметров.
* Подписать оси и заголовок
* Создать легенду
* Сохранить изображение в svg файл

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(array: np.array, alpha: float, beta: float):
    fxs = []
    for element in array:
        if element != 0:
            fx = (element**beta + alpha**beta) / element**beta
            fxs.append(fx)
        else:
            fxs.append(np.inf)
    return np.array(fxs)
```

## Задача 1

### Часть 1

Построить в общих осях графики для:
* α=1, β=1
* α=2, β=1
* α=1, β=2

```python
fig, ax = plt.subplots(figsize=(8, 7))
alphas = [1, 2, 1]
betas = [1, 1, 2]
x = np.hstack([np.linspace(-4, 0, 50), np.linspace(0, 4, 50)])
for alpha, beta in zip(alphas, betas):
    ax.plot(x, f(x, alpha, beta), label=f'$\\alpha=${alpha}, $\\beta=${beta},')

ax.set_xlim(-4, 4)
ax.set_ylim(-20, 20)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('f(x)', fontsize=20)
ax.set_title(r'График функции $f(x)=\frac{x^\beta+\alpha^\beta}{x^\beta}$ при различных $\alpha$, $\beta$', fontsize=20)
ax.legend(fontsize=15)
```

```python
fig.savefig("LearnFunction_task1.svg")
```

### Часть 2

Построить в общих осях графики для x>0.

На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:
* для малых x
* для больших x

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.

```python
fig, ax = plt.subplots(figsize=(8, 7))
alphas = [1, 2, 1]
betas = [1, 1, 2]
colors = ['r', 'g', 'b']
x = np.linspace(0, 10, 150)
for alpha, beta, color in zip(alphas, betas, colors):
    ax.plot(x, f(x, alpha, beta), c=color, label=f'$\\alpha=${alpha}, $\\beta=${beta},')

ax.set_xlim(0, 4)
ax.set_ylim(0, 20)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('f(x)', fontsize=20)
ax.set_title(r'График функции $f(x)=\frac{x^\beta+\alpha^\beta}{x^\beta}$ при различных $\alpha$, $\beta$', fontsize=20)
ax.legend(fontsize=15)

ax1 = fig.add_axes([0.28, 0.58, 0.3, 0.25])
x1 = np.linspace(0, 4, 200)
for alpha, beta, color in zip(alphas, betas, colors):
    ax1.plot(x1, f(x1, alpha, beta), c=color, label=f'$\\alpha=${alpha}, $\\beta=${beta},')
ax1.set_xlim(0, 4)
ax1.set_ylim(0, 5)
ax1.set_xlabel('x', fontsize=11)
ax1.set_ylabel('f(x)', fontsize=11)
ax1.set_title('Поведение функции f(x) при малых x', fontsize=10)

ax2 = fig.add_axes([0.55, 0.26, 0.3, 0.25])
x2 = np.linspace(100, 10000, 200)
for alpha, beta, color in zip(alphas, betas, colors):
    ax2.plot(x2, f(x2, alpha, beta), c=color, label=f'$\\alpha=${alpha}, $\\beta=${beta},')
ax2.set_xlim(100, 10000)
ax2.set_xlabel('x', fontsize=11)
ax2.set_ylabel('f(x)', fontsize=11)
ax2.set_title('Поведение функции f(x) при больших x', fontsize=10)
```

```python
fig.savefig("LearnFunction_task2.svg")
```

### Часть 3

Построить в общих осях графики для x<0.

На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к −∞.

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. Так же нанесите на графики прямую f(x) = 0.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений α и β.

```python
fig, ax = plt.subplots(figsize=(8, 7))
alphas = [1, 2, 1]
betas = [1, 1, 2]
colors = ['r', 'g', 'b']
x = np.linspace(-5, 0, 50)
for alpha, beta, color in zip(alphas, betas, colors):
    ax.plot(x, f(x, alpha, beta), c=color, label=f'$\\alpha=${alpha}, $\\beta=${beta},')
ax.plot(x, np.zeros_like(x), c='k', label='f(x)=0')
ax.set_xlim(-5, 0)
ax.set_ylim(-10, 10)
ax.set_yticks(range(-10, 10, 1))
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('f(x)', fontsize=20)
ax.set_title(r'График функции $f(x)=\frac{x^\beta+\alpha^\beta}{x^\beta}$ при различных $\alpha$, $\beta$', fontsize=20)
ax.legend(fontsize=15)

ax1 = fig.add_axes([0.25, 0.2, 0.4, 0.25])
x1 = np.linspace(-100000, 0, 300)
for alpha, beta, color in zip(alphas, betas, colors):
    ax1.plot(x1, f(x1, alpha, beta), c=color, label=f'$\\alpha=${alpha}, $\\beta=${beta},')
ax1.set_xlim(-100000, -2000)
ax1.set_ylim(0.9995, 1.0001)
ax1.set_xlabel('x', fontsize=11)
ax1.set_ylabel('f(x)', fontsize=11)
ax1.set_title(r'Поведение функции f(x) при удалении x к $-\infty$', fontsize=10)
```

```python
fig.savefig("LearnFunction_task3.svg")
```
