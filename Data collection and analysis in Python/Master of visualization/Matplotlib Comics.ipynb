Цель задания - составить комикс из нескольких изображений из папки for_comics

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('for_comics/1.png', 'rb') as read_file:
    map_img1 = plt.imread(read_file)

with open('for_comics/2.png', 'rb') as read_file:
    map_img2 = plt.imread(read_file)

with open('for_comics/3.png', 'rb') as read_file:
    map_img3 = plt.imread(read_file)

with open('for_comics/4.png', 'rb') as read_file:
    map_img4 = plt.imread(read_file)
```

```python
maps = [map_img1, map_img2, map_img3, map_img4]
```

```python
for map in maps:
    print(map.shape)
```

```python
fig = plt.figure(figsize=(10, 12))
plt.subplots_adjust(wspace=0.06, hspace=0.05)

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 3, 4)
ax3 = fig.add_subplot(2, 3, 5)
ax4 = fig.add_subplot(2, 3, 6)

axes = [ax1, ax2, ax3, ax4]

for ax, map_img in zip(axes, maps):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.imshow(map_img)
```

```python
fig.savefig("comics.svg")
```
