## –ó–∞–¥–∞–Ω–∏–µ 1

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∏ 2 —Ñ—É–Ω–∫—Ü–∏–π:

f(x) = sin(x)

–î–ª—è x –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ [-2ùúã; 2ùúã]

* –û—Ç–º–µ—Ç—å—Ç–µ –æ—Ç–¥–µ–ª—å–Ω—ã–º–∏ –º–∞—Ä–∫–µ—Ä–∞–º–∏ —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤. –û–±–∞ –≥—Ä–∞—Ñ–∏–∫–∞ –¥–æ–ª–∂–Ω—ã –ø—Ä–æ—Ö–æ–¥–∏—Ç—å —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –ø—Ä–∏ –ª—é–±–æ–º –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–∏
* –ü–æ–¥–ø–∏—à–∏—Ç–µ –æ—Å–∏ –∏ –≤—ã–≤–µ–¥–∏—Ç–µ –∑–∞–≥–æ–ª–æ–≤–æ–∫
* –í—ã–≤–µ–¥–∏—Ç–µ –æ—Å–µ–≤—ã–µ –ª–∏–Ω–∏–∏
* –ü–æ–¥–ø–∏—à–∏—Ç–µ –≥—Ä–∞—Ñ–∏–∫–∏

```python
def f(x):
    return np.sin(x)

def g(x):
    return 1 - np.cos(x)

x = np.linspace(-2*np.pi, 2*np.pi, 100)
```

```python
def x1(n):
    return np.pi/2 + 2*np.pi*n

def x2(n):
    return 2*np.pi*n

n_list = np.arange(-2, 2)
xn = np.hstack([x1(n_list), x2(n_list)])
f_g_xn = f(xn)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=f(x), mode='lines', name='f(x)=sin(x)'))
fig.add_trace(go.Scatter(x=x, y=g(x), mode='lines', name='g(x)=1-cos(x)'))
fig.add_trace(go.Scatter(x=xn, y=f_g_xn, mode='markers', name='–¢–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è',
                                                marker=dict(color="Green", size=7)))

fig.update_xaxes(range=[-2*np.pi, 2*np.pi], zeroline=True, zerolinewidth=1, zerolinecolor='Black')
fig.update_yaxes(zeroline=True, zerolinewidth=1, zerolinecolor='Black')

fig.update_layout(
    legend_orientation="h",
    margin=dict(l=0, r=0, t=30, b=0),
    title="–ì—Ä–∞—Ñ–∏–∫",
    xaxis_title="x",
    yaxis_title="y",
    hovermode='x'
)
fig.show()
```
